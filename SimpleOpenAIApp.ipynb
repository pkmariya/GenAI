```python
from openai import OpenAI
```

```python
client = OpenAI(
    api_key ="OPENAI_API_KEY"
)
```

```python
question = "What is the capital of India?"
```

```python
response = client.completions.create(
                model="gpt-4o-mini",
                temperature=0.7,
                top_p=0.3,
                prompt=question,
                # max_tokens=500
            )
```

```python
print(response.choices[0].text.strip())
```
